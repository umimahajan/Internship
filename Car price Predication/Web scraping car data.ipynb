```python
#Creating empty list 
Brand4 = []
Model4 = []
Variant4=[]
Manufacturing_Year4 = []
Driven_KiloMeters4= []
Fuel4 = []
Number_of_Owners4 = []

Car_Price4= []

#Fetching data for Brand    


#Fetching data for Model   
try:
    Model_name=driver.find_elements(By.XPATH,"//h2[@class='_3FpCg']")
    for i in Model_name:
        Model4.append(i.text)
except NoSuchElementException as e:
    Model4.append('-')

#Fetching data for Variant   
try:
    Variant_name=driver.find_elements(By.XPATH,"//p[@class='cvakB']")
    for i in Variant_name:
        Variant4.append(i.text)
except NoSuchElementException as e:
    Variant4.append('-')

#Fetching data for Manufacturing    
try:
    Manufacturing=driver.find_elements(By.XPATH,"//h2[@class='_3FpCg']")
    for i in Manufacturing:
        Manufacturing_Year4.append(i.text.split(" ")[0])
except NoSuchElementException as e:
    Manufacturing_Year4.append('-')    

#Fetching data for Driven_KiloMeters    
try:
    Driven_KiloMeter=driver.find_elements(By.XPATH,'//ul[@class="bVR0c"]/li[1]')
    for i in Driven_KiloMeter:
        Driven_KiloMeters4.append(i.text)
except NoSuchElementException as e:
    Driven_KiloMeters4.append('-')    

#Fetching data for Fuel    
try:
    Fuel_name=driver.find_elements(By.XPATH,'//ul[@class="bVR0c"]/li[3]')
    for i in Fuel_name:
        Fuel4.append(i.text)
except NoSuchElementException as e:
    Fuel4.append('-')

#Fetching data for Number_of_Owners    
try:
    Number_of_Owner=driver.find_elements(By.XPATH,'//ul[@class="bVR0c"]/li[2]')
    for i in Number_of_Owner:
        Number_of_Owners4.append(i.text)
except NoSuchElementException as e:
    Number_of_Owners4.append('-')


#Fetching data for Car_Price    
try:
    Car_Pr=driver.find_elements(By.XPATH,"//div[@class='_7udZZ']/span[1]")
    for i in Car_Pr:
        Car_Price4.append(i.text)
except NoSuchElementException as e:
    Car_Price4.append('-')    
```

```python
Brandm=[]
try:
    Brand_name=driver.find_elements(By.XPATH,"//h2[@class='_3FpCg']")
    for i in Brand_name:
        Brandm.append(i.text.split(" ")[1])
except NoSuchElementException as e:
    Brandm.append('-')
```

```python
len(Brandm),len(Model4),len(Variant4),len(Manufacturing_Year4),len(Driven_KiloMeters4),len(Fuel4),len(Number_of_Owners4),len(Car_Price4)
```
